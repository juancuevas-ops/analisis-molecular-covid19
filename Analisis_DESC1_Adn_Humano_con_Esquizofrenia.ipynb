```python
import matplotlib.pyplot as plt
from Bio import SeqIO

sequences = []

# Open the FASTA file
with open("/content/protein_esquiz.faa", "r") as handle:
    # Parse the FASTA records
    for record in SeqIO.parse(handle, "fasta"):
        sequences.append(record.seq)

# Extract sequence lengths for plotting
sequence_lengths = [len(seq) for seq in sequences]

# Create a scatter plot using sequence lengths
plt.figure(figsize=(10, 6))
plt.scatter(range(len(sequence_lengths)), sequence_lengths)  # Use sequence_lengths instead of sequences
plt.xlabel("Sequence Index")
plt.ylabel("Sequence Length")
plt.title("Scatter Plot of Sequence Lengths")
plt.show()
```

```python
#  ver la sequencia de aminoacidos en proteina

# Extract the protein sequence from the first record
protein_sequence = next(SeqIO.parse("/content/protein_esquiz.faa", "fasta")).seq

# Print the protein sequence
print(protein_sequence)
```

```python
#  grafico ´para ver los aminoacidos

from Bio.SeqUtils import ProtParam

# Calculate various physicochemical properties of the protein sequence
protein_properties = ProtParam.ProteinAnalysis(str(protein_sequence))

# Access specific properties
molecular_weight = protein_properties.molecular_weight()
isoelectric_point = protein_properties.isoelectric_point()

# Print the results
print(f"Molecular weight: {molecular_weight}")
print(f"Isoelectric point: {isoelectric_point}")

# Generate a bar chart of amino acid composition
# Use get_amino_acids_percent() instead of amino_acid_content
```


# Análisis de datos moleculares
Peso molecular: 91122.7562 g/mol

Interpretación:

Basándome en los datos moleculares proporcionados, no puedo ofrecer un diagnóstico o explicación precisa sin información adicional sobre la proteína en cuestión. Sin embargo, puedo ofrecer algunas interpretaciones generales sobre lo que estos valores podrían indicar.

Peso molecular:

El peso molecular de una proteína es un indicador de su tamaño y complejidad. Un peso molecular elevado, como el de 91122.7562 g/mol, sugiere que la proteína es grande y compleja. Esto podría implicar que la proteína tiene múltiples subunidades o que contiene dominios proteicos de gran tamaño.

Punto isoeléctrico:

El punto isoeléctrico (pI) es el pH en el que una proteína tiene una carga neta neutra. Un pI de 6.0048 sugiere que la proteína es ligeramente ácida. Esto es consistente con la idea de que la proteína contiene muchos residuos de aminoácidos ácidos, como el ácido aspártico y el ácido glutámico.

```python
import matplotlib.pyplot as plt
```

```python
#  (grafico de barras de la composicion de aminoacidos)

amino_acid_composition = protein_properties.get_amino_acids_percent()

# Prepare data for the bar chart
labels = list(amino_acid_composition.keys())
values = list(amino_acid_composition.values())

# Plot the bar chart

plt.bar(labels, values)
plt.xlabel("Amino Acid")
plt.ylabel("% Composition")
plt.title("Amino Acid Composition of Protein")
plt.show()
```
