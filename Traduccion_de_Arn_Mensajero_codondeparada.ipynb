```python
from Bio.Seq import Seq
messenger_rna = Seq("AUGGCCAUUGUAAUGGGCCGCUGAAAGGGUGCCCGAUAG")
messenger_rna
Seq('AUGGCCAUUGUAAUGGGCCGCUGAAAGGGUGCCCGAUAG') #SECUENCIA DE BASES NITROGENADAS DE ARN MENSAJERO
```

Los ribosomas son responsables de la síntesis de proteínas, en un proceso conocido como traducción. La información necesaria para esa síntesis se encuentra en el ARN mensajero (ARN), cuya secuencia de nucleótidos, determina la secuencia de aminoácidos de la proteína.ahora podemos divisar gracias a algoritmos de IA.este sorprendente momento

```python
messenger_rna.translate()
```

AHORA LOGRAREMOS EL MISMO PROCESO MOLECULAR PERO DIRECTAMENTE DEL ADN SIN HABER SIDO TRADUCIDO A ARNm

```python
from Bio.Seq import Seq 
coding_dna = Seq ( "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG") 
coding_dna 
Seq ( 'ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG') 
coding_dna.translate () 
Seq ( 'MAIVMGR * KGAR *')
```

el resultado de cadena de aminoacidos es el mismo ya que dentro de la funcion de traduccion se incluye la de trancripcion

```python
#Ahora, es posible que desee traducir los nucleótidos hasta el primer codón de parada del marco y luego detenerse (como sucede en la naturaleza):

coding_dna.translate () 
Seq ('MAIVMGR * KGAR *') 
```

```python
coding_dna.translate (to_stop = True) #demo de codones de parada
Seq ('MAIVMGR') 
```

```python
coding_dna.translate (to_stop = True)# el codon de parada en si solo se muestra como astrisco (se puede modificar esa funcion)
Seq ('MAIVMGRWKGAR * ') 
```

```python
coding_dna.translate ( to_stop = True) #tabla 2
Seq (' MAIVMGRWKGAR ')
```
